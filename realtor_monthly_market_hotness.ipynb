### 1) Read Realtor Monthly Market Hotness by County
[Data Location](https://www.realtor.com/research/data/)

```python
%%time
# read in county data
url = 'https://econdata.s3-us-west-2.amazonaws.com/Reports/Hotness/RDC_Inventory_Hotness_Metrics_County_History.csv'
_df = pd.read_csv(url)
print(f'Num of records: {len(_df)}')
print(f'Num of columns: {len(_df.columns)}')
```

## <font color="blue">Transformations</font>

```python
# view dataset by removing last row
_df.iloc[:-1,:].tail(1)
```

### 1) Features

```python
df_features = _df.iloc[:-1,:].copy()
df_features['state'] = df_features.apply(lambda x:
  x['county_name'].split(', ')[-1].upper(), axis=1)
df_features['time_period'] = df_features.apply(lambda x:
  datetime.strptime(str(x['month_date_yyyymm']), '%Y%m'), axis=1)
df_features['hotness_state_rank'] = df_features.groupby(['state', 'time_period'])['hotness_score'].rank()
df_features.head(1)
```

### 2) Filter

```python
# state of interest
state = 'FL'
```

```python
df_filter = df_features.loc[df_features['state'] == state]
print(f'Num of records: {len(df_filter)}')
print(f'Num of columns: {len(df_filter.columns)}')
print('Num of unq counties:', len(df_filter['county_name'].unique()))
print('Num of unq time period:', len(df_filter['month_date_yyyymm'].unique()))
df_filter.head()
```

```python
print('Min time period:', df_filter['time_period'].min())
print('Max time period:', df_filter['time_period'].max())
```

```python
# hottest top 10 markets in recent time period
df_top_10_hotness = df_filter.loc[df_filter['time_period'] == df_filter['time_period'].max()]\
  .sort_values(by=['hotness_state_rank']).head(10).reset_index(drop=True)
df_top_10_hotness
```

## <font color="blue">Visualizations</font>

```python
# get top 10 counties
top_10_counties = df_top_10_hotness['county_name'].tolist()
top_10_counties_fips = df_top_10_hotness['county_fips'].tolist()
top_10_counties
```

```python
# hotness score
df_plot = df_filter.loc[df_filter['county_name'].isin(top_10_counties)]
fig = px.line(df_plot, x="time_period", y="hotness_score", color='county_name', title=f'Hottest markets in <b>{state}</b>')
fig.show()
```

```python
# median days on market
df_plot = df_filter.loc[df_filter['county_name'].isin(top_10_counties)]
fig = px.line(df_plot, x="time_period", y="median_days_on_market", color='county_name', title=f'Hottest markets in <b>{state}</b>')
fig.show()
```

```python
# listing price
df_plot = df_filter.loc[df_filter['county_name'].isin(top_10_counties)]
fig = px.line(df_plot, x="time_period", y="median_listing_price", color='county_name', title=f'Hottest markets in <b>{state}</b>')
fig.show()
```
